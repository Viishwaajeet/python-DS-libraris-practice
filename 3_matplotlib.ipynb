# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
```

```python
b = a*2
```

```python
x = np.arange(0,11)
```

```python
y = x**2
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
type(axes)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x,y)
ax[1].plot(a,b)
```

```python
type(ax)
```

```python
ax
```

```python
s = np.linspace(0,15,100)
fig,ax = plt.subplots(nrows=3, ncols=1)
ax[0].plot(s,np.sin(s))
ax[1].plot(s,np.cos(s))
ax[2].plot(s,np.tan(s))

plt.tight_layout()
```

```python
s = np.linspace(0,15,100)
fig,ax = plt.subplots(nrows=3,ncols=2)

ax[0][0].plot(s,np.sin(s))
ax[0][0].set_xlabel('sin')
ax[0][1].plot(s,np.arcsin(s))  #cosec
ax[0][1].set_xlabel('cosec')
ax[1][0].plot(s,np.cos(s))
ax[1][0].set_xlabel('cos')
ax[1][1].plot(s,np.arccos(s))  #sec
ax[1][1].set_xlabel('sec')
ax[2][0].plot(s,np.tan(s))
ax[2][0].set_xlabel('tan')
ax[2][1].plot(s,np.arctan(s))  #cot
ax[2][1].set_xlabel('cot')

fig.suptitle("Trigonometric function graphs", fontsize=20)

plt.tight_layout()
# fig.subplots_adjust(wspace=0.25, hspace=0.4)

fig.savefig('mythirdplot.png',bbox_inches='tight')
```

# styling

```python
fig = plt.figure()
x = np.linspace(0,10,11)
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x, label="X vs X")
axes.plot(x, x**2, label="X vs X^2")

axes.legend(loc=0)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,11,10)
```

```python
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color='#245c31',label='x vs x',ls='-.')   #using hex code
ax.plot(x,x+1, color="lightblue", label='x vs x+1', lw=10, ls='--')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x, color='#245c31',label='x vs x',lw=5)   #using hex code

lines[0].set_dashes([5,2,10,2])

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, marker='o',ms=10,color='#245c31',label='x vs x',ls='--',
       markerfacecolor='green',markeredgewidth=4,markeredgecolor='lightblue')   #using hex code
ax.plot(x,x+1, marker='P',markersize=10,color='lightblue',label='x vs x+1')   #using hex code

ax.legend()
```
