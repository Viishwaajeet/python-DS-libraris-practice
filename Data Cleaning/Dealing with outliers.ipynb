```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_ages(mu=50, sigma=13, num_samples=100, seed=42):
    np.random.seed(seed)
    sample_ages = np.random.normal(loc=mu,scale=sigma,size=num_samples)
    sample_ages = np.round(sample_ages,decimals=0)
    return sample_ages
```

```python
sample = create_ages()
```

```python
sample
```

```python
sns.displot(data=sample,bins=20,kde=True)
```

```python
sns.boxplot(data=sample,orient='h')
```

```python
ser = pd.Series(sample)
```

```python
ser.describe()
```

```python
ser[ser>18]
```

```python
def fix_values(age):
    if age<18:
        return 18
    else:
        return age
```

```python
ser = ser.apply(fix_values)
```

```python
ser[ser<18]
```
