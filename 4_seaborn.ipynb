# Categorical Plots

```python
df.head()
```

```python
plt.figure(dpi=200)
sns.countplot(data=df,x='division')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(dpi=200)
sns.countplot(data=df,x='level of education', hue='division')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=300)
sns.barplot(data=df,x='level of education',y='salary',
           estimator=np.mean, errorbar='sd',hue='division')

plt.legend(bbox_to_anchor=(0.25,1.3))
```
